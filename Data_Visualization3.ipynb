```
import matplotlib.pyplot as plt
import pandas as pd
```

```
df = pd.read_csv('bank.csv')
df
```

```
plt.title("BANK DATA")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.plot(df['Gender'] , df['Age'],color='green', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=12)
plt.grid()
plt.legend()
plt.show()
```

```
plt.title("BANK INFO")
plt.xlabel("Geography")
plt.ylabel("NumOfProducts")
plt.bar(df["Geography"] , df["NumOfProducts"] , color = ("cyan" , "yellow") ,width = 0.8)

plt.show()
```

```
plt.title("BANK INFO ")
plt.xlabel("NumOfProducts")
plt.ylabel("EstimatedSalary")
plt.scatter(df['NumOfProducts'],df['EstimatedSalary'] , color = 'brown' , marker= "*")
plt.show()
```

```
plt.xlabel('HasCrCard')
plt.ylabel('IsActiveMember')
plt.grid()
plt.scatter(df['HasCrCard'],df['IsActiveMember'] , color ="red" , marker = "*")
plt.show()
```

```
plt.xlabel('Age')
plt.ylabel('Exited')
plt.grid()
plt.pie(df['Age'],df['Exited'] )
plt.show()
```

```
plt.xlabel('CreditScore')
plt.ylabel('Balance')
plt.grid()
plt.scatter(df['CreditScore'],df['Balance'] , marker ="*")
plt.show()
```
