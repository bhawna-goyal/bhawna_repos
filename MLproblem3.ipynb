```
import matplotlib.pyplot as plt
import pandas as pd
```

```
df = pd.read_csv("student.csv")
df
colors = ("yellow","beige","cyan","pink")
```

```
plt.title("STUDENTS REPORT")
# draw pie chart......
 #explode basically to making a single or multiple wedge.
  #wedgeprop is basically used to give wedge properties like edgecolor and linewidth etc.
plt.pie(df.iloc[0:,1]  ,labels = df.iloc[0:,0] , explode = (0,0,0.1,0.2) ,autopct = "%1.1f%%" , shadow = True , startangle=90 ,colors = colors,
       wedgeprops   = { 'linewidth' : 5,'edgecolor' : "black" })  
plt.legend()
plt.axis('equal')
plt.show()
```
